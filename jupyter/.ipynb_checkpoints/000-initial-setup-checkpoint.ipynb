# Convert CSV into MySQL DB with sqlalchemy

### Create MySQL docker:
```yaml
version: "3.9"
services:
  #############################################
  ####           mariadb Server           #####
  ##############################################
  mysql:
    image: mysql:latest
    container_name: nyahua-mysql
    environment:
      MYSQL_ROOT_PASSWORD: 9002005
      MYSQL_DATABASE: nyahuaDB
      MYSQL_USER: nyahua
      MYSQL_PASSWORD: 123456
    ports:
      - "3306:3306"
    volumes:
      - ./mariadb:/var/lib/mysql
```

### database management
-  you can enter into the container: `docker exec -it nyahua-mysql bash` then `mysql -u username -p`
- `enter` as root and grant all rights to `nyahua`
```bash
karibu@nyahua:~$ docker exec -it nyahua-mysql mysql -u root -p
mysql> GRANT ALL PRIVILEGES ON *.* TO 'nyahua'@'%';
```
- create your database
```bash
docker exec -it nyahua-container mysql -u nyahua -p
```


````python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base

# MySQL database configuration
db_config = {
    'user': 'nyahua',
    'password': '123456',
    'host': 'nyahua.com',
    'port': 3306,
    'database': 'movie'
}

# Create a database connection
database_url = f"mysql+mysqlconnector://{db_config['user']}:{db_config['password']}@{db_config['host']}:{db_config['port']}/{db_config['database']}"
engine = create_engine(database_url)
    
````

````python
# CSV file path
csv_file_path = '../dataset/movie.csv'
df = pd.read_csv(csv_file_path)

table_name = 'movies'
df.to_sql(table_name, con=engine, index=False, if_exists='replace')
````

````python
with engine.connect() as connection:
    # Perform database operations here
    result = connection.execute("SELECT * FROM movies LIMIT 100;")
    
pd.DataFrame(result)
````
